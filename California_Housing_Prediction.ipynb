## Import the Libaries and Prepare the dataset


```python
## Import the libraries
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
import matplotlib.pyplot as plt
```

```python
## Prepare the data
dataset = fetch_california_housing()
print(dataset.DESCR)
```

```python
print(dataset)
```

## Simple Linear Regression with one feature

```python
## Simple Linear Regression
x = dataset['data'][...,0:1]
y = dataset['target']
print(x.shape,y.shape)
```

```python
plt.scatter(x[...,0:1],y)
plt.xlabel('MedInc')
plt.ylabel('MedHouseVal')
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20)

model = LinearRegression()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)
```

```python
plt.scatter(x[...,0:1],y)
plt.plot(x_test,y_pred, color='red')
plt.xlabel('MedInc')
plt.ylabel('MedHouseVal')
```

```python
w = model.coef_
b = model.intercept_
print(w,b)
```

```python
print(y_pred[0], w*x_test[0]+b)
```

```python
r2_scr = r2_score(y_test, y_pred)
mse_error = mean_squared_error(y_test, y_pred)

print(r2_scr, mse_error)
```

## Multi Linear Regression

```python
## Split the test train data with 20% size selection
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20)

## Select the model and fit the data
model = LinearRegression()
model.fit(x_train, y_train)

## Predict with the test data
y_pred = model.predict(x_test)
```

```python
## Get the model coefficient and intercept values
w = model.coef_
b = model.intercept_
print(w,b)
```

```python
## Quantitative Metrics,  R2 score and MSE error
r2_scr = r2_score(y_test, y_pred)
mse_error = mean_squared_error(y_test, y_pred)

print(r2_scr, mse_error)
```
